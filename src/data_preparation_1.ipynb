```python
indian = []
categories = list(question1['categories'])
for c in categories:
    i = 0
    if 'Indian' in c:
        i = 1
    indian.append(i)
print(sum(indian))

question1['indian'] = indian
question1.head(5)
```

```python
question1_1A = question1[question1['neighborhood'].isin(['Scarborough','Etobicoke'])]
question1_1A = question1_1A[(question1_1A['is_open'] == 1) & (question1_1A['indian'] == 1)]
question1_1A = question1_1A[["business_id","name","neighborhood","review_count","stars"]]
print(len(question1_1A),len(question1_1A[question1_1A['neighborhood'] == 'Scarborough']),len(question1_1A[question1_1A['neighborhood'] == 'Etobicoke']))
question1_1A.to_csv("data/question1_1A.csv",index=False)
question1_1A.head()
```

```python
review = pd.read_csv("data/reviews.csv")
review.head(1)
```

```python
IDs = question1_1A['business_id']
question1_1B = review[review['business_id'].isin(IDs)]
question1_1B = pd.merge(question1_1B[['business_id','review_id','stars']],question1_1A[['business_id','name','neighborhood']])

print(len(question1_1B),len(question1_1B[question1_1B['neighborhood']=='Scarborough']),
     len(question1_1B[question1_1B['neighborhood']=='Etobicoke']))

question1_1B.to_csv("data/question1_1B.csv",index=False)
question1_1B.head()
```
